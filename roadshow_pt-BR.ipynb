<H1><B>Vector Database</B></H1>

Bancos de dados vetoriais são um tipo especializado de banco de dados projetado para armazenar e consultar vetores, que são sequências de números que representam características ou atributos de dados.

Ao invés de fazer buscas baseadas em valor (keywords), o banco vetorial permite fazer buscas por semântica (significado).

- <B>Exemplo de busca por valor</B>: Retorne os clientes com idade acima de 20 anos

- <B>Exemplo de busca por semântica</B>: Retorne 3 clientes mais similares ao perfil da Jessica

Como um exemplo didático (apenas didático para entender a intuição), considere o seguinte exemplo:

````python
import database as db
db.init()
db.get_customers()
````

No exemplo acima, temos 4 colunas:

```
Customer (Id, Name, Age, Income)
```

Vamos plotar um gráfico com esses clientes, onde cada ponto representa um vetor de duas dimensões: 

```
[AGE, INCOME]
```

````python
db.customers_chart("age,income").show()
````

O eixo X representa a idade (age), e o eixo Y representa a renda (income). Cada cliente é posicionado no gráfico em função da combinação da idade com a renda. 

Ao observar o gráfico, é possível perceber que <b>Joshua</b> e <b>David</b> são os clientes "*mais semelhantes*" ao <b>James</b>. Vamos ver se isso é verdade realizando uma busca por similaridade:

````python
db.similarity_chart("age,income", "James").show()
````

Conforme esperado, Joshua e David (em negrito) estãos mais próximos de James. A similaridade é obtida através do uso de cálculos que envolvem distância entre vetores. Existem várias métricas de distância. As mais conhecidas e utilizadas são EUCLIDEAN e COSINE.
